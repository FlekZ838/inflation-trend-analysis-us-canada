```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt

# 下载美国 CPI 数据（从 FRED）
us_cpi = data.DataReader("CPIAUCSL", "fred", start="2010-01-01")
us_cpi.reset_index(inplace=True)
us_cpi.columns = ['Date', 'CPI_US']

# 计算同比增长率
us_cpi['YoY_US'] = us_cpi['CPI_US'].pct_change(periods=12) * 100

# 绘制趋势图
plt.figure(figsize=(12,6))
plt.plot(us_cpi['Date'], us_cpi['YoY_US'], label='US YoY CPI')
plt.axhline(2, color='gray', linestyle='--', label='Target Inflation')
plt.title("Year-over-Year Inflation (US)")
plt.xlabel("Date")
plt.ylabel("Inflation (%)")
plt.legend()
plt.grid(True)
plt.show()
```
